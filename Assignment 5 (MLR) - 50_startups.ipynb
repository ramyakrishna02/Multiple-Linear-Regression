## Fitting the model

The Equation would be,

Profit = B0 + B1(RDS) + B2(AD) + B3(MS) + e

```python
model = smf.ols('Profit ~ RDS+AD+MS',data=startups).fit()
model.summary()
```

### Simple Linear Regression Models

```python
# model fitting for the Administration independent variable 
# The equation would be -- Profit = B0 + B1(AD) + e

AD_m = smf.ols('Profit~AD',data=startups).fit()
print(AD_m.pvalues)

# to know its significance 
if AD_m.pvalues['AD'] < 0.05:
    print('\n AD is significant variable.')
else:
    print('\n AD is insignificant variable.')
```

```python
# model fitting for the Marketing Spend independent variable 
# The equation would be -- Profit = B0 + B1(MS) + e

MS_m = smf.ols('Profit~MS',data=startups).fit()
print(MS_m.pvalues)

# to know its significance
if MS_m.pvalues['MS'] < 0.05:
    print('\n MS is significant variable.')
else:
    print('\n MS is insignificant variable.')
```

```python
# model fitting for Administration and Marketing Spend variables 
# The equation would be -- Profit = B0 + B1(AD) + B2(MS) + everything 

ADMS_m = smf.ols('Profit~AD+MS',data=startups).fit()
print(ADMS_m.pvalues)

# to know their significance 
if all(ADMS_m.pvalues[['AD','MS']] < 0.05):
    print('\n Both are significant variables.')
elif any(ADMS_m.pvalues[['AD','MS']] < 0.05):
    if ADMS_m.pvalues['AD'] < 0.05:
        print('\n AD is a significant variable and MS is an insignificant variable.')
    else:
        print('\n MS is a significant variable and AD is an insignificant variable.')
else:
    print('\n Both are insignificant variables.')
```

### Calculating VIF 
    Variance Inflation Factor

```python
rsq_AD = smf.ols('AD ~ RDS+MS',data=startups).fit().rsquared
print('VIF of AD : ',1/(1-rsq_AD))

rsq_MS = smf.ols('MS ~ RDS+AD',data=startups).fit().rsquared
print('VIF of MS : ',1/(1-rsq_MS))
```

## Residual Analysis 

### Test for Normality of Residuals (Q-Q Plot)

```python
# qq-plot for residuals
import statsmodels.api as sm
qq = sm.qqplot(model.resid,line='q')
plt.title('Normal Q-Q plot of residuals')
plt.show()
```

### Residual Plot for Homoscedasticity

```python
# creating function to get standardized values of residuals and fitted values
def get_standardized_values(vals):
    return (vals - vals.mean())/vals.std()
```

```python
# scatter plot on the standardized values of residuals and fitted values
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized Residual values')
plt.show()
```

## Model Deletion Diagnostics 

#### Detecting Influencers/Outliers



### Cook's Distance method

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
# Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(startups)), np.round(c,3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
# index and value of influencer where c is more than others
(np.argmax(c), np.max(c))
```

### Leverage Value method

#### High Influence points

```python
from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.show()
```

```python
# to get leverage cutoff value
k = 3        # no. of independent variables
n = startups.shape[0]
leverage_cutoff = 3*((k+1)/n)
leverage_cutoff
```

```python
# to identify the outlier beyond the leverage cutoff value

influence_plot(model)
x = [leverage_cutoff,leverage_cutoff]
y = [-5,3]
plt.plot(x, y,color='darkred', linewidth=2)
plt.show()
```

##### From the above plot, it is evident that data points 46 and 49 are the influencers

```python
startups[startups.index.isin([46,49])]
```

## Improving the model

```python
# printing the original data
startups.head()
```

```python
# Discard the data points which are influencers and reasign the row number (reset_index())
startups1 = startups.drop(startups.index[[46,49]],axis=0).reset_index()
```

```python
# Drop the original index
startups1 = startups1.drop(['index'],axis=1)

# printing the new data after removing outliers and index
startups1
```

## Build Model

```python
#Exclude variable "AD" and generate R-squared, Adjusted R-sq, AIC and BIC values
final_MS = smf.ols('Profit ~ RDS+MS',data=startups1).fit()
```

```python
print('R-sq value: ',final_MS.rsquared,
      '\nAdj R-sq value: ',final_MS.rsquared_adj,
      '\nAIC: ',final_MS.aic,'\nBIC: ',final_MS.bic)
```

```python
# Exclude variable "MS" and generate R-squared, Adjusted R-sq, AIC and BIC values
final_AD = smf.ols('Profit ~ RDS+AD',data=startups1).fit()
```

```python
print('R-sq value: ',final_AD.rsquared,
      '\nAdj R-sq value: ',final_AD.rsquared_adj,
      '\nAIC: ',final_AD.aic,'\nBIC: ',final_AD.bic)
```

#### Comparing above R-sq, adj r-sq, AIC and BIC values, the final model without 'Administration' variable has high r-sq, adj r-sq and low AIC, BIC values.

#### Hence, consider the model with 'Marketing Spend' variable.

## Model Deletion Diagnostics and Final Model

```python
# Further process for identifying influencers by cook's distance method and finalising the best fit model
startups2 = startups1.copy()

while np.max(c)>0.5 :
    model_influence_MS = final_MS.get_influence()
    (c_MS,_)=model_influence_MS.cooks_distance
    np.argmax(c_MS), np.max(c_MS)
    
    startups2 = startups1.drop(startups1.index[[np.argmax(c_MS)]],axis=0).reset_index(drop=True)    
    startups2
    
else:
    final_model=smf.ols("Profit ~ RDS+MS",data=startups2).fit()
    
    print('R-sq value: ',final_model.rsquared,
          '\nAdj R-sq value: ',final_model.rsquared_adj,
          '\nAIC: ',final_model.aic,'\nBIC: ',final_model.bic)
```
